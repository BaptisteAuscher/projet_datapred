```python
from get_from_news_api import *
import itertools
import pandas as pd
```

```python
energy_keywords = [ "power", "electricity", "natural gaz", "energy" ]

countries = [ 
    "Russia", "Norway", "Netherlands", "Germany", "United Kingdom", "France", "Italy", "Belgium", \
    "Spain","Denmark","Poland","Ukraine","Turkey","Austria","Hungary", "Czech Republic", "Slovakia", \
    "Sweden", "Finland", "Lithuania" 
]

cartesian_product = [list(item) for item in itertools.product(energy_keywords, countries)]

acronyms = [
    ["Gazprom"], ["GTSOU"], ["Gassco"], ["GRTgaz"], ["Eni"], ["Snam"], ["National Grid"], ["Gasunie"], ["OMV"], 
    ["Fluxys"], ["Equinor"], ["PGNiG"], ["Uniper"], ["RWE"], ["Engie"], ["Enag√°s"], ["BP"], ["TotalEnergies"], 
    ["Naftogaz"], ["Nord Stream AG"]
]

names = [
    ["Gazprom"], ["Gas Transmission System Operator of Ukraine"], ["Gassco"], ["Gaz R√©seau Transport France"], 
    ["Ente Nazionale Idrocarburi"], ["Snam S.p.A."], ["National Grid plc"], ["N.V. Nederlandse Gasunie"], ["OMV AG"], 
    ["Fluxys Belgium"], ["Equinor ASA"], ["Polskie G√≥rnictwo Naftowe i Gazownictwo"], ["Uniper SE"], ["RWE AG"], 
    ["Engie S.A."], ["Enag√°s S.A."], ["British Petroleum"], ["TotalEnergies SE"], ["Naftogaz of Ukraine"], 
    ["Nord Stream AG"], ["trading hub europe"], ["Title Transfer Facility"], ["TTF", "Netherlands"], 
    ["Point d'√âchange de Gaz"], ["PSV", "Italy"]
]

keywords = cartesian_product + acronyms + names         
```

```python
time_interval = ['2024-05-01', '2024-05-31'] # Curieusement, il ne trouve rien quand je met une date trop ancienne (i.e. il y a quelques mois...)
max_rows = 1000
```

```python
result = [ get_articles(keywords=sublist, dateStart=time_interval[0], dateEnd=time_interval[1], keywordsLoc="body", andOrKW="and", maxRows=max_rows) \
            for sublist in keywords ]
```

```python
# Enregistrer les dataframes au format csv
k = 1
for res in result :
    res.to_csv(f"../corpus/corpus_{k}.csv", index=False)
    k += 1
```
